<h1>Preprocessing data</h1>

```python
from sklearn.preprocessing import Normalizer
X = train_data.drop(['target'],axis=1)
y = pd.get_dummies(train_data['target'])
```

```python
def preprocessing_CategoricalInput(X):
    X = X.drop(['num_outbound_cmds'],axis = 1)
    object_columns = X.dtypes[X.dtypes == 'object'].index.tolist()
    new_X = pd.get_dummies(X, columns = object_columns, drop_first = True)
    return new_X

new_X = preprocessing_CategoricalInput(X)
```

```python
nor = Normalizer().fit(new_X)
new_X = nor.transform(new_X)
def preprocessing_Input(X):
    new_X = preprocessing_CategoricalInput(X)
    new_X = nor.transform(new_X)
    return new_X
```

<h1>Split data</h1>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(new_X,y,test_size=0.2,random_state=42)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_train,y_train,test_size=0.1,random_state=12)
```

```python
print('Train shape: ',X_train.shape)
print('Valid shape: ',X_valid.shape)
print('Test shape: ',X_test.shape)
```

```python
num_input = X_train.shape[1]
num_input
```

```python
num_classes = y_train.shape[1]
num_classes
```

<h1>build and evaluate model</h1>

<h2>Model1</h2>

```python
# Creating a Neural Network Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout,BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint, ReduceLROnPlateau
import tensorflow as tf
```

```python
# having 19 neuron is based on the number of available features
model1 = Sequential()
model1.add(Dense(118,input_dim=num_input, kernel_initializer='normal',activation='relu'))
model1.add(Dense(100,activation='relu'))
model1.add(Dense(80,activation='relu'))
model1.add(Dense(60,activation='relu'))
model1.add(Dense(40,activation='relu'))
model1.add(Dense(20,activation='relu'))
model1.add(Dense(10,activation='relu'))
model1.add(Dense(num_classes,activation='softmax'))
model1.compile(optimizer='Adam',loss='CategoricalCrossentropy',metrics=['accuracy'])
model1.summary()
```

```python
checkpoint1 = ModelCheckpoint('./checkpoint/Ex2/Model1.h5',monitor='val_loss',save_best_only=True,save_weights_only=True,mode='min')
learning_rate1 = ReduceLROnPlateau(patience=5,verbose=1,factor=0.5,min_delta=0.000001)
early_stopping1 = EarlyStopping(patience=30,monitor='val_loss',verbose=1,restore_best_weights=True,mode='min')
```

```python
import time
pre_train1 = time.time()
history = model1.fit(X_train,y_train,validation_data=(X_valid,y_valid),
                    epochs=500,batch_size = 128,callbacks=[early_stopping1,learning_rate1,checkpoint1])
time_training1 = time.time() - pre_train1
```
